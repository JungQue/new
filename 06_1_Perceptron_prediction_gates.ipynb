### Perceptrons - Making Predictions
#### Creating a gate with Perceptron

```python
import numpy as np
```

### AND Gate

```python
def AND(x1, x2):
    x = np.array([x1, x2])#배열 생성
    w = np.array([0.5, 0.5])
    b = -0.5
    #activation 계산
    tmp = w[0]*x[0] + w[1]*x[1] + b 
    # tmp = np.sum(w*x) + b # 배열wx를 각 순서대로 곱하여 더함 그후 b값을 더하고 tmp에 저장
    #b값은 그래프의 평행이동 담당 
    print(tmp)
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
AND(1,0)
```

```python

AND(1,0)
```

```python
AND(0,0)
```

```python
AND(1,1)
```

### NAND Gate

```python
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5])
    b = 0.7
    tmp = w[0]*x[0] + w[1]*x[1] + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
NAND(0,0)
```

```python
NAND(1,0)
```

```python
NAND(0,1)
```

```python
NAND(1,1)
```

### OR Gate

```python
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
OR(0,0)
```

```python
OR(1,0)
```

```python
OR(0,1)
```

```python
OR(1,1)
```

### XOR Gate

```python
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y
```

```python
XOR(0,0)
```

```python
XOR(1,1)
```

```python
XOR(0,1)
```

```python
XOR(1,0)
```

XOR cannot be expressed as a single layer Perceptron.
